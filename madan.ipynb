```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('DataSet.csv')
data.size
```

```python
data.head()
```

```python
data.describe()
```

```python
data = data.dropna()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(data['Year'], data['No_Smoothing'])
plt.xlabel('Year')
plt.ylabel('No_Smoothing')
plt.title('No_Smoothing over Time')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Year'], data['Lowess(5)'])
plt.xlabel('Year')
plt.ylabel('Lowess(5)')
plt.title('Lowess(5) over Time')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data[['No_Smoothing', 'Lowess(5)']].values)
plt.xticks([1, 2], ['No_Smoothing', 'Lowess(5)'])
plt.ylabel('Value')
plt.title('Comparison of No_Smoothing and Lowess(5)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.stackplot(data['Year'], data[['No_Smoothing', 'Lowess(5)']].values.T, labels=['No_Smoothing', 'Lowess(5)'])
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Comparison of No_Smoothing and Lowess(5)')
plt.legend()
plt.show()
```
