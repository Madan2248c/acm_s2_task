```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('DataSet.csv')
data.size
```

```python
data.head()
```

```python
data.describe()
```

```python
data = data.dropna()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(data['Year'], data['No_Smoothing'])
plt.xlabel('Year')
plt.ylabel('No_Smoothing')
plt.title('No_Smoothing over Years')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Year'], data['Lowess(5)'])
plt.xlabel('Year')
plt.ylabel('Lowess(5)')
plt.title('Lowess(5) over Years')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data['No_Smoothing'])
plt.ylabel('No_Smoothing')
plt.title('No_Smoothing Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Lowess(5)'], bins=10)
plt.xlabel('Lowess(5)')
plt.ylabel('Frequency')
plt.title('Lowess(5) Distribution')
plt.show()
```
